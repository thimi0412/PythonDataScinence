## matplotlibの基礎

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
```

```python
ax = plt.axes()
```

* figureクラスは軸、グラフィックス、テキスト、ラベルを含む1つのコンテナ
* axesクラスはメモリとラベルを持つ境界

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x))
```

```python
# 複数回呼び出せばグラフを追加できる
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

### スタイル

```python
plt.plot(x, np.sin(x), color='blue') #色名
plt.plot(x, np.sin(x-1), color='g') # tgbcmykのいずれか
plt.plot(x, np.sin(x-2), color='0.75') # グレイスケール 0~1
plt.plot(x, np.sin(x-3), color='#FFDD44') # カラーコード
plt.plot(x, np.sin(x-4), color=(1.0, 0.2, 0.3)) # RGB
```

```python
plt.plot(x, np.sqrt(x), linestyle='solid') # 直線(デフォルト)
plt.plot(x, np.sqrt(x)+0.5, linestyle='dashed')  # 破線
plt.plot(x, np.sqrt(x)+1.0, linestyle='dashdot') #一点鎖線
plt.plot(x, np.sqrt(x)+1.5, linestyle='dotted') # 点線 
```

### 座標軸の範囲

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

```python
# axisを使用する方法
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

### ラベル付け

```python
plt.plot(x, np.sin(x))
plt.title('Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
plt.plot(x, np.sin(x), color='blue', label='sin(x)')
plt.plot(x, np.cos(x), color='red', label='cos(x)')
plt.axis('equal')
plt.legend()
```
